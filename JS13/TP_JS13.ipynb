```python
# %% [markdown]
# # Tugas Praktikum: Klasifikasi Angka Tulisan Tangan (MNIST) dengan JST

# %%
# Import libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

print("TensorFlow version:", tf.__version__)

# %% [markdown]
# ## BAGIAN 1: LOAD DAN EKSPLORASI DATASET MNIST

# %%
print("="*70)
print("BAGIAN 1: LOAD DAN EKSPLORASI DATASET MNIST")
print("="*70)

# Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"\nUkuran data training: {X_train.shape}")
print(f"Ukuran label training: {y_train.shape}")
print(f"Ukuran data testing: {X_test.shape}")
print(f"Ukuran label testing: {y_test.shape}")
print(f"\nRentang nilai pixel: {X_train.min()} - {X_train.max()}")
print(f"Jumlah kelas: {len(np.unique(y_train))}")

# %%
# Visualisasi beberapa contoh gambar
plt.figure(figsize=(12, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.suptitle("Contoh Dataset MNIST", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# %%
# Normalisasi data (0-255 → 0-1)
X_train_normalized = X_train / 255.0
X_test_normalized = X_test / 255.0

# One-hot encoding label
y_train_categorical = to_categorical(y_train, 10)
y_test_categorical = to_categorical(y_test, 10)

print("Setelah preprocessing:")
print(f"Rentang nilai pixel: {X_train_normalized.min()} - {X_train_normalized.max()}")
print(f"Shape label setelah one-hot encoding: {y_train_categorical.shape}")
print(f"Contoh label: {y_train[0]} → {y_train_categorical[0]}")

# %% [markdown]
# ## BAGIAN 2: MODEL DASAR (2 HIDDEN LAYERS)

# %%
print("="*70)
print("BAGIAN 2: MODEL DASAR (2 HIDDEN LAYERS)")
print("="*70)

# Bangun model JST dengan 2 hidden layer
model_dasar = Sequential([
    Flatten(input_shape=(28, 28)),  # Ubah gambar 28x28 menjadi vektor 784
    Dense(128, activation='relu'),   # Hidden layer 1
    Dense(64, activation='relu'),    # Hidden layer 2
    Dense(10, activation='softmax')  # Output layer (10 kelas)
], name='Model_Dasar')

# Kompilasi model
model_dasar.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\n=== Arsitektur Model Dasar ===")
model_dasar.summary()

# %%
# Latih model
print("\nMelatih model dasar...")
start_time = time.time()

history_dasar = model_dasar.fit(
    X_train_normalized, y_train_categorical,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=1
)

training_time_dasar = time.time() - start_time
print(f"\nWaktu pelatihan: {training_time_dasar:.2f} detik")

# %%
# Evaluasi model
loss_dasar, acc_dasar = model_dasar.evaluate(X_test_normalized, y_test_categorical, verbose=0)
print(f"\n=== Hasil Evaluasi Model Dasar ===")
print(f"Loss pada data uji: {loss_dasar:.4f}")
print(f"Akurasi pada data uji: {acc_dasar:.4f} ({acc_dasar*100:.2f}%)")

# %%
# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_dasar.history['loss'], label='Train Loss')
plt.plot(history_dasar.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Model Dasar - Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_dasar.history['accuracy'], label='Train Accuracy')
plt.plot(history_dasar.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Model Dasar - Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## BAGIAN 3: EKSPERIMEN DENGAN BERBAGAI KONFIGURASI

# %%
print("="*70)
print("BAGIAN 3: EKSPERIMEN DENGAN BERBAGAI KONFIGURASI")
print("="*70)

# Dictionary untuk menyimpan hasil eksperimen
results = []

# %% [markdown]
# ### EKSPERIMEN 1: Ubah jumlah neuron (256 dan 128)

# %%
print("\n--- EKSPERIMEN 1: Neuron lebih banyak (256, 128) ---")

model_exp1 = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
], name='Model_256_128')

model_exp1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp1.summary()

start_time = time.time()
history_exp1 = model_exp1.fit(
    X_train_normalized, y_train_categorical,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=0
)
training_time_exp1 = time.time() - start_time

loss_exp1, acc_exp1 = model_exp1.evaluate(X_test_normalized, y_test_categorical, verbose=0)
print(f"Akurasi: {acc_exp1:.4f}, Waktu: {training_time_exp1:.2f}s")

results.append({
    'Model': 'Neuron 256-128',
    'Neurons': '256-128',
    'Layers': 2,
    'Activation': 'ReLU',
    'Accuracy': acc_exp1,
    'Loss': loss_exp1,
    'Time (s)': training_time_exp1
})

# %% [markdown]
# ### EKSPERIMEN 2: Tambah 1 hidden layer (128, 64, 32)

# %%
print("\n--- EKSPERIMEN 2: Tambah 1 hidden layer (128, 64, 32) ---")

model_exp2 = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
], name='Model_3_Layers')

model_exp2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_exp2.summary()

start_time = time.time()
history_exp2 = model_exp2.fit(
    X_train_normalized, y_train_categorical,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=0
)
training_time_exp2 = time.time() - start_time

loss_exp2, acc_exp2 = model_exp2.evaluate(X_test_normalized, y_test_categorical, verbose=0)
print(f"Akurasi: {acc_exp2:.4f}, Waktu: {training_time_exp2:.2f}s")

results.append({
    'Model': '3 Hidden Layers',
    'Neurons': '128-64-32',
    'Layers': 3,
    'Activation': 'ReLU',
    'Accuracy': acc_exp2,
    'Loss': loss_exp2,
    'Time (s)': training_time_exp2
})

# %% [markdown]
# ### EKSPERIMEN 3: Fungsi aktivasi Sigmoid vs ReLU

# %%
print("\n--- EKSPERIMEN 3: Fungsi aktivasi Sigmoid ---")

model_sigmoid = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='sigmoid'),
    Dense(64, activation='sigmoid'),
    Dense(10, activation='softmax')
], name='Model_Sigmoid')

model_sigmoid.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time = time.time()
history_sigmoid = model_sigmoid.fit(
    X_train_normalized, y_train_categorical,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=0
)
training_time_sigmoid = time.time() - start_time

loss_sigmoid, acc_sigmoid = model_sigmoid.evaluate(X_test_normalized, y_test_categorical, verbose=0)
print(f"Akurasi: {acc_sigmoid:.4f}, Waktu: {training_time_sigmoid:.2f}s")

results.append({
    'Model': 'Sigmoid Activation',
    'Neurons': '128-64',
    'Layers': 2,
    'Activation': 'Sigmoid',
    'Accuracy': acc_sigmoid,
    'Loss': loss_sigmoid,
    'Time (s)': training_time_sigmoid
})

# %% [markdown]
# ### EKSPERIMEN 4: Neuron lebih sedikit (64, 32)

# %%
print("\n--- EKSPERIMEN 4: Neuron lebih sedikit (64, 32) ---")

model_small = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
], name='Model_64_32')

model_small.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

start_time = time.time()
history_small = model_small.fit(
    X_train_normalized, y_train_categorical,
    validation_split=0.2,
    epochs=10,
    batch_size=128,
    verbose=0
)
training_time_small = time.time() - start_time

loss_small, acc_small = model_small.evaluate(X_test_normalized, y_test_categorical, verbose=0)
print(f"Akurasi: {acc_small:.4f}, Waktu: {training_time_small:.2f}s")

results.append({
    'Model': 'Neuron 64-32',
    'Neurons': '64-32',
    'Layers': 2,
    'Activation': 'ReLU',
    'Accuracy': acc_small,
    'Loss': loss_small,
    'Time (s)': training_time_small
})

# %%
# Tambahkan model dasar ke hasil
results.insert(0, {
    'Model': 'Model Dasar',
    'Neurons': '128-64',
    'Layers': 2,
    'Activation': 'ReLU',
    'Accuracy': acc_dasar,
    'Loss': loss_dasar,
    'Time (s)': training_time_dasar
})

# %% [markdown]
# ## BAGIAN 4: PERBANDINGAN HASIL SEMUA EKSPERIMEN

# %%
print("="*70)
print("BAGIAN 4: PERBANDINGAN HASIL SEMUA EKSPERIMEN")
print("="*70)

# Buat dataframe hasil
df_results = pd.DataFrame(results)
print("\n")
print(df_results.to_string(index=False))

# %%
# Visualisasi perbandingan
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Perbandingan Akurasi
axes[0, 0].barh(df_results['Model'], df_results['Accuracy'], color='skyblue')
axes[0, 0].set_xlabel('Accuracy')
axes[0, 0].set_title('Perbandingan Akurasi')
axes[0, 0].set_xlim([0.9, 1.0])
axes[0, 0].grid(axis='x', alpha=0.3)
for i, v in enumerate(df_results['Accuracy']):
    axes[0, 0].text(v, i, f' {v:.4f}', va='center')

# 2. Perbandingan Waktu Pelatihan
axes[0, 1].barh(df_results['Model'], df_results['Time (s)'], color='lightcoral')
axes[0, 1].set_xlabel('Waktu (detik)')
axes[0, 1].set_title('Perbandingan Waktu Pelatihan')
axes[0, 1].grid(axis='x', alpha=0.3)
for i, v in enumerate(df_results['Time (s)']):
    axes[0, 1].text(v, i, f' {v:.1f}s', va='center')

# 3. Perbandingan Loss
axes[1, 0].barh(df_results['Model'], df_results['Loss'], color='lightgreen')
axes[1, 0].set_xlabel('Loss')
axes[1, 0].set_title('Perbandingan Loss')
axes[1, 0].grid(axis='x', alpha=0.3)
for i, v in enumerate(df_results['Loss']):
    axes[1, 0].text(v, i, f' {v:.4f}', va='center')

# 4. Accuracy vs Training Time
axes[1, 1].scatter(df_results['Time (s)'], df_results['Accuracy'], 
                   s=200, alpha=0.6, c=range(len(df_results)), cmap='viridis')
for i, model in enumerate(df_results['Model']):
    axes[1, 1].annotate(model, 
                       (df_results['Time (s)'][i], df_results['Accuracy'][i]),
                       fontsize=8, ha='right')
axes[1, 1].set_xlabel('Waktu Pelatihan (detik)')
axes[1, 1].set_ylabel('Accuracy')
axes[1, 1].set_title('Akurasi vs Waktu Pelatihan')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## BAGIAN 5: ANALISIS DAN KESIMPULAN

# %%
print("="*70)
print("BAGIAN 5: ANALISIS DAN KESIMPULAN")
print("="*70)

best_acc_idx = df_results['Accuracy'].idxmax()
fastest_idx = df_results['Time (s)'].idxmin()

print(f"\n✓ Model dengan akurasi tertinggi: {df_results.loc[best_acc_idx, 'Model']}")
print(f"  Akurasi: {df_results.loc[best_acc_idx, 'Accuracy']:.4f}")
print(f"  Waktu: {df_results.loc[best_acc_idx, 'Time (s)']:.2f}s")

print(f"\n✓ Model tercepat: {df_results.loc[fastest_idx, 'Model']}")
print(f"  Akurasi: {df_results.loc[fastest_idx, 'Accuracy']:.4f}")
print(f"  Waktu: {df_results.loc[fastest_idx, 'Time (s)']:.2f}s")

print("\n📊 KESIMPULAN:")
print("1. ReLU umumnya memberikan performa lebih baik dibanding Sigmoid")
print("2. Menambah neuron meningkatkan akurasi tapi juga waktu training")
print("3. Menambah layer tidak selalu meningkatkan akurasi secara signifikan")
print("4. Trade-off antara akurasi dan kecepatan perlu dipertimbangkan")

# %% [markdown]
# ## BAGIAN 6: VISUALISASI PREDIKSI MODEL TERBAIK

# %%
print("="*70)
print("BAGIAN 6: VISUALISASI PREDIKSI MODEL TERBAIK")
print("="*70)

# Gunakan model dengan akurasi terbaik
best_model = model_dasar  # Bisa diganti dengan model lain

# Prediksi pada data test
predictions = best_model.predict(X_test_normalized[:20], verbose=0)
predicted_labels = np.argmax(predictions, axis=1)
true_labels = y_test[:20]

# %%
# Visualisasi prediksi
plt.figure(figsize=(15, 6))
for i in range(20):
    plt.subplot(4, 5, i+1)
    plt.imshow(X_test[i], cmap='gray')
    
    color = 'green' if predicted_labels[i] == true_labels[i] else 'red'
    plt.title(f"Pred: {predicted_labels[i]}\nTrue: {true_labels[i]}", 
             color=color, fontsize=9)
    plt.axis('off')

plt.suptitle("Prediksi Model (Hijau=Benar, Merah=Salah)", fontsize=14, y=1.00)
plt.tight_layout()
plt.show()

# %%
# Confusion Matrix
y_pred_all = np.argmax(best_model.predict(X_test_normalized, verbose=0), axis=1)
cm = confusion_matrix(y_test, y_pred_all)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Model Terbaik')
plt.show()

# %%
# Classification Report
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred_all, target_names=[str(i) for i in range(10)]))

# %%
print("="*70)
print("PRAKTIKUM SELESAI!")
print("="*70)
```
