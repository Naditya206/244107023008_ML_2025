Deskripsi Tugas

Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

  1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
  2. Lakukan proses encoding pada kolom "diagnosis".
  3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('wbc.csv')
df.head(100)
df.shape
df.describe()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
# Variabel yang tidak digunakan
unused_cols = ["id", "Unnamed: 32"]

# Variabel yang digunakan
used_cols = [col for col in df.columns if col not in unused_cols]

print("Variabel yang tidak digunakan:", unused_cols)
print("Variabel yang digunakan:", used_cols)
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

# Encoding diagnosis (M=1, B=0)
encoder = LabelEncoder()
df["diagnosis"] = encoder.fit_transform(df["diagnosis"])

print(df[["diagnosis"]].head(10))
```

3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler

# Standardisasi semua kolom numerik (selain diagnosis)
scaler = StandardScaler()
numeric_cols = df.drop(columns=["diagnosis", "id", "Unnamed: 32"]).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Hasil akhir
print(df.head())
```
