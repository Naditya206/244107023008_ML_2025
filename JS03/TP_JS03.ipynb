1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
import pandas as pd

df = pd.read_csv('sample_data/wbc.csv')

# Pisahkan variabel yang dapat digunakan dan yang tidak
# Kolom 'id' tidak diperlukan untuk analisis, jadi kita hapus
df = df.drop(columns='id')

# Tampilkan informasi singkat mengenai data untuk verifikasi
print("Informasi data setelah menghapus kolom 'id':")
print(df.info())
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Lakukan encoding pada kolom 'diagnosis'
df['diagnosis_encoded'] = le.fit_transform(df['diagnosis'])

# Tampilkan 5 baris pertama untuk melihat hasil encoding
print("\nData setelah encoding pada kolom 'diagnosis':")
print(df[['diagnosis', 'diagnosis_encoded']].head())

# Hapus kolom 'diagnosis' yang asli
df = df.drop(columns='diagnosis')
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler

# Pisahkan variabel fitur (X) dan target (y)
X = df.drop(columns='diagnosis_encoded')
y = df['diagnosis_encoded']

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Lakukan standarisasi pada semua kolom numerik di X
X_scaled = scaler.fit_transform(X)

# Buat kembali X_scaled menjadi DataFrame dengan nama kolom yang benar
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\nData fitur (X) setelah standarisasi:")
print(X_scaled.head())
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split

# Lakukan stratified split dengan rasio 80:20
# 'random_state' digunakan agar hasil split selalu sama
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Cetak dimensi (shape) dari setiap dataset untuk verifikasi
print("\nDimensi data latih dan uji:")
print(f"Bentuk (shape) X_train: {X_train.shape}")
print(f"Bentuk (shape) X_test: {X_test.shape}")
print(f"Bentuk (shape) y_train: {y_train.shape}")
print(f"Bentuk (shape) y_test: {y_test.shape}")
```
